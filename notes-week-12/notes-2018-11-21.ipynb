```python
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt
%matplotlib inline
```

## RLC Circuit

Let's plot solutions of

\begin{align*}
\dot{v}_a &= \frac{-2}{RC} v_a + \frac{1}{RC}v_b \\
\dot{v}_b &= \frac{1}{RC} v_a - \frac{2}{RC}v_b
\end{align*}

```python
R = 1; C = 1;
t = np.linspace(0,1,100)
M = np.array([[1,1],[-1,1]])
v0 = np.array([[1],[-1]])
coeffs = la.solve(M,v0)
va = coeffs[0,0]*np.exp(-3*t/R/C) + coeffs[1,0]*np.exp(-t/R/C)
vb = -coeffs[0,0]*np.exp(-3*t/R/C) + coeffs[1,0]*np.exp(-t/R/C)
plt.plot(t,va,t,vb)
plt.legend(['$v_a$','$v_b$'])
plt.show()
```

```python
R = 1; C = 1;
t = np.linspace(0,10,200)
M = np.array([[1,1],[-1,1]])
N = 100
for n in range(0,N):
    theta = 2*np.pi*n/N
    v0 = 2*np.array([[np.cos(theta)],[np.sin(theta)]])
    coeffs = la.solve(M,v0)
    va = coeffs[0,0]*np.exp(-3*t/R/C) + coeffs[1,0]*np.exp(-t/R/C)
    vb = -coeffs[0,0]*np.exp(-3*t/R/C) + coeffs[1,0]*np.exp(-t/R/C)
    plt.plot(va,vb,'b')
plt.grid(True), plt.axis('equal')
plt.xlim([-1,1]), plt.ylim([-1,1])
plt.show()
```
