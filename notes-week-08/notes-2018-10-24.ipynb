## Central Difference Formula

The derivative $f'(x)$ of a function $f(x)$ at $x=a$ is the limit

$$
f'(a) = \lim_{h \to 0} \frac{f(a+h) - f(a)}{h}
$$

The central difference formula is the average of the forward difference and the backward difference formulas

$$
f'(a) \approx \frac{1}{2} \left( \frac{f(a+h) - f(a)}{h} + \frac{f(a) - f(a-h)}{h} \right) = \frac{f(a+h) - f(a-h)}{2h}
$$

Write a function called `derivative` which takes 3 input parameters `f`, `a` and `h` (default value `h=0.001`) and return the central difference formula

$$
f'(a) \approx \frac{f(a+h) - f(a-h)}{2h}
$$

```python
def derivative(f,a,h=0.001):
    return (f(a+h) - f(a-h))/(2*h)
```

Let's test on a simple function:

```python
derivative(lambda x: x**2,1)
```

Notice that our function is vectorized (when `f` is a vectorized function). Therefore we can input an array of values for `a`.

```python
x = np.linspace(0,5*np.pi,100)
y = derivative(np.cos,x)
plt.plot(x,y)
plt.show()
```

## Examples

Use `derivative` to plot $f'(x)$ where

$$
f(x) = \left( \frac{4x^2 + 2x + 1}{x + 2e^x} \right)^x
$$

```python
f = lambda x: ((4*x**2 + 2*x + 1)/(x + 2*np.exp(x)))**x
x = np.linspace(0,6,100)
y = f(x)
dfdx = derivative(f,x)
plt.plot(x,dfdx,x,y), plt.grid(True), plt.axis('equal')
plt.show()
```

Use `derivative` to plot $f'(x)$ where

$$
f(x) = \sqrt{x + \sqrt{x + \sqrt{x}}}
$$

Notice that we can't evaluate $f(x)$ when $x$ is negative therefore we need to compute starting at $x > h = 0.001$.

```python
f = lambda x: np.sqrt(x + np.sqrt(x + np.sqrt(x)))
x = np.linspace(0.1,10,100)
y = f(x)
dfdx = derivative(f,x)
plt.plot(x,dfdx,x,y), plt.grid(True)
plt.axis('equal'), plt.legend(["f'(x)",'f(x)'])
plt.show()
```
