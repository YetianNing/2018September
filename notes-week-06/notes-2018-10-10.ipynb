## Plotting Procedure

The basic procedure to plot the graph of a function $y=f(x)$ is:

1. Create an array of `x` values, for example using `numpy.linspace`.
2. Create an array of `y` values using array operations and functions.
3. Enter `plt.plot(x,y)`.
4. Add style using `pyplot` commands.
5. End with `plt.show()`.

For example, let's plot $y = \sin(x) + 0.5\cos(20x)$ for $x \in [0,5\pi]$.

```python
x = np.linspace(0,5*np.pi,1000)
y = np.sin(x) + 0.5*np.cos(20*x)
plt.plot(x,y)
plt.show()
```

## Pyplot Commands

We can use `pyplot` commands to add:

* title
* labels
* grid
* colors,
* legend
* line and marker styles

Let's plot the degrees 2, 4, and 6 Taylor polynomials of $y=\cos(x)$ for $x \in [-2\pi,2\pi]$. Include a title and a legend.

\begin{align}
T_2(x) &= 1 - \frac{x^2}{2!} \\
T_4(x) &= 1 - \frac{x^2}{2!} + \frac{x^4}{4!} \\
T_6(x) &= 1 - \frac{x^2}{2!} + \frac{x^4}{4!} - \frac{x^6}{6!}  \\
\end{align}

```python
a = -2*np.pi; b = 2*np.pi;
x = np.linspace(a,b,100)
y = np.cos(x)
y1 = 1 - x**2/2
y2 = 1 - x**2/2 + x**4/np.math.factorial(4)
y3 = 1 - x**2/2 + x**4/np.math.factorial(4) - x**6/np.math.factorial(6)
plt.plot(x,y,'--',x,y1,'k:',x,y2,'-.',x,y3,':')
plt.ylim([-3,3]), plt.xlim([a,b])
plt.legend(['cos(x)','Degree 2','Degree 4','Degree 6'],loc='lower right')
plt.title('Taylor Polynomials for cos(x)')
plt.grid(True), plt.xlabel('X Values'), plt.ylabel('Y Values')
plt.show()
```
